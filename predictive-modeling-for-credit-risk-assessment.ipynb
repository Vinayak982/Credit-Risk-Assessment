```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/kaggle/input/credit-risk-analysis/credit_risk.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print(data.info())

# Display the first few rows of the dataset
print(data.head())

# Check the distribution of the target variable
print(data['Default'].value_counts())

# Preprocess the data
# Convert categorical columns to dummy variables
data = pd.get_dummies(data, columns=['Home', 'Intent', 'Status'], drop_first=True)

# Encode 'Default' column as binary
data['Default'] = data['Default'].apply(lambda x: 1 if x == 'Yes' else 0)

# Define features (X) and target (y)
X = data.drop(['Id', 'Default'], axis=1)
y = data['Default']

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Split the data into training and testing sets using StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Standardize the feature data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize results dictionary to store evaluation metrics
results = {}

# Train and evaluate models
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    if len(np.unique(y_train)) < 2:
        print("---------")
        continue

    # Logistic Regression
    log_reg = LogisticRegression()
    log_reg.fit(X_train, y_train)
    y_pred_log_reg = log_reg.predict(X_test)
    results['Logistic Regression'] = (y_test, y_pred_log_reg)
    
    # Decision Tree Classifier
    tree_clf = DecisionTreeClassifier()
    tree_clf.fit(X_train, y_train)
    y_pred_tree = tree_clf.predict(X_test)
    results['Decision Tree'] = (y_test, y_pred_tree)
    
    # Random Forest Classifier
    rf_clf = RandomForestClassifier()
    rf_clf.fit(X_train, y_train)
    y_pred_rf = rf_clf.predict(X_test)
    results['Random Forest'] = (y_test, y_pred_rf)

# Evaluation Function
def evaluate_model(y_test, y_pred, model_name):
    print(f'--- {model_name} ---')
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    print('Classification Report:')
    print(classification_report(y_test, y_pred))
    print('Accuracy Score:', accuracy_score(y_test, y_pred))
    print('\n')

# Evaluate the models
for model_name, (y_test, y_pred) in results.items():
    evaluate_model(y_test, y_pred, model_name)

# Plotting Confusion Matrices
def plot_confusion_matrix(y_test, y_pred, model_name):
    plt.figure(figsize=(6, 4))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Default', 'Default'], yticklabels=['No Default', 'Default'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

# Plot confusion matrices for each model
for model_name, (y_test, y_pred) in results.items():
    plot_confusion_matrix(y_test, y_pred, model_name)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/kaggle/input/credit-risk-analysis/credit_risk.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print(data.info())

# Display the first few rows of the dataset
print(data.head())

# Check the distribution of the target variable
print(data['Default'].value_counts())

# Feature Engineering
# 1. Convert categorical columns to dummy variables
data = pd.get_dummies(data, columns=['Home', 'Intent', 'Status'], drop_first=True)

# 2. Encode 'Default' column as binary
data['Default'] = data['Default'].apply(lambda x: 1 if x == 'Y' else 0)

# 3. Create new feature: Income-to-loan ratio
data['Income_to_loan_ratio'] = data['Income'] / data['Amount']

# Handle missing values before creating polynomial features
imputer = SimpleImputer(strategy='mean')
data[['Rate', 'Income_to_loan_ratio']] = imputer.fit_transform(data[['Rate', 'Income_to_loan_ratio']])

# 4. Create polynomial features for 'Rate' and 'Income_to_loan_ratio'
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(data[['Rate', 'Income_to_loan_ratio']])
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(['Rate', 'Income_to_loan_ratio']))
data = pd.concat([data, poly_features_df], axis=1)

# Define features (X) and target (y)
X = data.drop(['Id', 'Default'], axis=1)
y = data['Default']

# Impute the remaining missing values for the entire dataset
X = imputer.fit_transform(X)

# Split the data into training and testing sets using StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Standardize the feature data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize results dictionary to store evaluation metrics
results = {}
final_predictions = {}

# Train and evaluate models
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    if len(np.unique(y_train)) < 2:
        print("Skipping this split as it does not contain both classes.")
        continue

    # Logistic Regression
    log_reg = LogisticRegression()
    log_reg.fit(X_train, y_train)
    y_pred_log_reg = log_reg.predict(X_test)
    results['Logistic Regression'] = (y_test, y_pred_log_reg)
    
    # Decision Tree Classifier
    tree_clf = DecisionTreeClassifier()
    tree_clf.fit(X_train, y_train)
    y_pred_tree = tree_clf.predict(X_test)
    results['Decision Tree'] = (y_test, y_pred_tree)
    
    # Random Forest Classifier
    rf_clf = RandomForestClassifier()
    rf_clf.fit(X_train, y_train)
    y_pred_rf = rf_clf.predict(X_test)
    results['Random Forest'] = (y_test, y_pred_rf)

# Evaluation Function
def evaluate_model(y_test, y_pred, model_name):
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    
    print(f'--- {model_name} ---')
    print('Confusion Matrix:')
    print(cm)
    print('Classification Report:')
    print(classification_report(y_test, y_pred))
    print(f'Accuracy Score: {acc:.4f}')
    print(f'Precision Score: {precision:.4f}')
    print(f'Recall Score: {recall:.4f}')
    print(f'F1 Score: {f1:.4f}')
    print('\n')
    
    return acc, precision, recall, f1, cm

# Store the final evaluation metrics
final_metrics = {}

# Evaluate the models
for model_name, (y_test, y_pred) in results.items():
    final_metrics[model_name] = evaluate_model(y_test, y_pred, model_name)

# Plotting Confusion Matrices
def plot_confusion_matrix(cm, model_name):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Default', 'Default'], yticklabels=['No Default', 'Default'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

# Plot confusion matrices for each model
for model_name, metrics in final_metrics.items():
    _, _, _, _, cm = metrics
    plot_confusion_matrix(cm, model_name)

# Generate and evaluate final predictions
for model_name, (y_test, y_pred) in final_predictions.items():
    print(f'--- Final Evaluation for {model_name} ---')
    evaluate_model(y_test, y_pred, model_name)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, ttest_ind

# Load the dataset
file_path = '/kaggle/input/credit-risk-analysis/credit_risk.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Feature Engineering
# 1. Convert categorical columns to dummy variables
data = pd.get_dummies(data, columns=['Home', 'Intent', 'Status'], drop_first=True)

# 2. Encode 'Default' column as binary
data['Default'] = data['Default'].apply(lambda x: 1 if x == 'Y' else 0)

# 3. Create new feature: Income-to-loan ratio
data['Income_to_loan_ratio'] = data['Income'] / data['Amount']

# Handle missing values before creating polynomial features
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
data[['Rate', 'Income_to_loan_ratio']] = imputer.fit_transform(data[['Rate', 'Income_to_loan_ratio']])

# 4. Create polynomial features for 'Rate' and 'Income_to_loan_ratio'
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(data[['Rate', 'Income_to_loan_ratio']])
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(['Rate', 'Income_to_loan_ratio']))
data = pd.concat([data, poly_features_df], axis=1)

# Statistical Tests
# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 2. Chi-squared test for categorical features
chi2_results = []
categorical_features = ['Home_RENT', 'Home_MORTGAGE', 'Intent_EDUCATION', 'Intent_MEDICAL', 'Intent_PERSONAL', 'Intent_DEBTCONSOLIDATION', 'Intent_HOMEIMPROVEMENT', 'Intent_VENTURE']
for feature in categorical_features:
    if feature in data.columns:
        contingency_table = pd.crosstab(data[feature], data['Default'])
        chi2, p, dof, ex = chi2_contingency(contingency_table)
        chi2_results.append({'Feature': feature, 'Chi2': chi2, 'p-value': p})

chi2_df = pd.DataFrame(chi2_results)

plt.figure(figsize=(10, 6))
sns.barplot(x='Feature', y='Chi2', data=chi2_df)
plt.title('Chi-squared Test Results for Categorical Features')
plt.xticks(rotation=45)
plt.show()
```
