# Credit Risk Assissment

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
df=pd.read_csv('credit_risk.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],kde=True, bins=30)
plt.title('Distribution of Age')
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.show


plt.figure(figsize=(10,6))
sns.histplot(df['Income'],kde=True, bins=30)
plt.title('Distribution of Income')
plt.ylabel('Frequency')
plt.xlabel('Income')
plt.show
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Emp_length'], kde=True, bins=30)
plt.title('Employment Length Distribution')
plt.ylabel('Frequency')
plt.xlabel('Emp_length')
plt.show


plt.figure(figsize=(10,6))
sns.countplot(x='Intent',data=df)
plt.xlabel('Intent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Amount'], kde=True, bins=30)
plt.title('Loan Amount Distribution')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Rate'], kde=True, bins=30)
plt.title('Interest Rate Distribution')
plt.xlabel('Interest Rate')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(x='Status', data=df)
plt.title('Loan Approval Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(x='Default', data=df)
plt.title('Loan Default Status')
plt.xlabel('Default')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Cred_length'], kde=True, bins=30)
plt.title('Credit Length Distribution')
plt.xlabel('Credit Length (years)')
plt.ylabel('Frequency')
plt.show()

```
